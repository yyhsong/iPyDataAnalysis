```python
import numpy as np
import pandas as pd
```

# X和Y之间的相关性分析

- X增大，Y增大，两个变量正相关
- X增大，Y减小，两个变量负相关
- X增大，Y无视，两个变量不相关

## 协方差

- 协方差>0，X和Y正相关
- 协方差<0，X和Y负相关
- 协方差=0，X和Y独立无关

## Pearson相关系数[-1, 1]

- 0.8 - 1.0  极强相关
- 0.6 - 0.8  强相关
- 0.4 - 0.6  中等程度相关
- 0.2 - 0.4  弱相关
- 0.0 - 0.2  极弱相关或无相关

# 相关分析函数

- cov()  计算协方差矩阵
- corr()  计算相关系数矩阵，Pearson、Spearman、Kendall等系数

```python
# 示例：房价增幅与M2增幅的相关性
hprice = pd.Series([3.04, 22.93, 12.75, 22.6, 12.33], index=['2008', '2009', '2010', '2011', '2012'])
m2 = pd.Series([8.18, 18.38, 9.13, 15.82, 6.69], index=['2008', '2009', '2010', '2011', '2012'])

# 计算相关系数
hprice.corr(m2)
```

```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['font.family'] = 'Microsoft YaHei'
matplotlib.rcParams['font.size'] = 14
```

```python
years = ['2008', '2009', '2010', '2011', '2012']
hprice = [3.04, 22.93, 12.75, 22.6, 12.33]
m2 = [8.18, 18.38, 9.13, 15.82, 6.69]

plt.plot(years, hprice, label='房价增幅')
plt.plot(years, m2, label='M2增幅')

plt.title('房价增幅与M2增幅的相关性')
plt.legend(loc='best')
# plt.grid(True)
plt.show()
```
