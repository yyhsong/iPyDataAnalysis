```python
import pandas as pd
```

```python
df = pd.read_csv('./data/titanic.csv')
df
```

```python
# 根据索引取特定的值
df[['Name', 'Sex', 'Age']][:5]
```

```python
# 用iloc来通过索引定位数据
df.iloc[0]
df.iloc[0:5]
df.iloc[0:5, 0:6]
```

```python
# 用loc来通过label定位数据
df = df.set_index('Name') # 把Name定义为索引 
df.loc['Heikkinen, Miss. Laina']
```

```python
df.loc['Heikkinen, Miss. Laina', 'Age']
```

```python
df.loc['Heikkinen, Miss. Laina':'Allen, Mr. William Henry', :]
```

```python
df.loc['Heikkinen, Miss. Laina','Fare'] = 1000
df.head()
```

```python
# 布尔类型的索引
df['Age'] > 30
```

```python
df[df['Age'] > 30][:5]
```

```python
df[df['Sex'] == 'male'][:5]
```
