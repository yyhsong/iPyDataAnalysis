```python
import pandas as pd
```

```python
df = pd.DataFrame({'Month': ["January", "January", "January", "January", 
                                  "February", "February", "February", "February", 
                                  "March", "March", "March", "March"],
                   'Category': ["Transportation", "Grocery", "Household", "Entertainment",
                                "Transportation", "Grocery", "Household", "Entertainment",
                                "Transportation", "Grocery", "Household", "Entertainment"],
                   'Amount': [74., 235., 175., 100., 115., 240., 225., 125., 90., 260., 200., 120.]})
df
```

```python
# 数据透视
df_pivot = df.pivot(index='Category', columns='Month', values='Amount')
df_pivot
```

```python
# 按列求和
df_pivot.sum()
```

```python
# 按行求和
df_pivot.sum(axis=1)
```

```python
df = pd.read_csv('./data/titanic.csv')
df
```

```python
# 数据透视表，默认求平均值
df.pivot_table(index='Sex', columns='Pclass', values='Fare')
```

```python
# 求最大值
df.pivot_table(index='Sex', columns='Pclass', values='Fare', aggfunc='max')
```

```python
# 统计个数
df.pivot_table(index='Sex', columns='Survived', values='Fare', aggfunc='count')
```

```python
# 统计个数
pd.crosstab(index=df['Sex'], columns=df['Survived'])
```
