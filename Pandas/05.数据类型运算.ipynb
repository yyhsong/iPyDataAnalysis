```python
import numpy as np
import pandas as pd
```

# 算术运算法则

- 算术运算根据行列索引，补齐后运算，运算默认产生浮点数
- 补齐时缺项填充NaN
- 二维和一维，一维和零维间为广播运算
- 采用+ - * /符号进行的二元运算产生新的对象

```python
a = pd.DataFrame(np.arange(12).reshape(3, 4))

b = pd.DataFrame(np.arange(20).reshape(4, 5))

# 自动补齐NaN
a + b
```

## 方法形式的运算

- add(d, **argws)  类型间加法运算
- sub(d, **argws)  类型间减法运算
- mul(d, **argws)  类型间乘法运算
- div(d, **argws)  类型间除法运算

```python
a = pd.DataFrame(np.arange(12).reshape(3, 4))

b = pd.DataFrame(np.arange(20).reshape(4, 5))

# 指定填充数据代替NaN
a.mul(b, fill_value=0)
```

```python
a = pd.DataFrame(np.arange(20).reshape(4, 5))
b = pd.Series(np.arange(4))

# 不同维度间为广播运算
# 一维Series默认在轴axis=1（行）参与运算
a - b

# 使用运算方法可以使一维Series在轴axis=0（列）参与运算
a.sub(b, axis=0)
```

# 比较运算法则

- 比较运算只能比较相同索引的元素，不进行补齐
- 二维和一维，一维和零维间为广播运算
- 采用 > < >= <= == != 等符号进行的二元运算产生布尔对象

```python
a = pd.DataFrame(np.arange(12).reshape(3, 4))
b = pd.DataFrame(np.arange(12, 0, -1).reshape(3, 4))

# 同纬度运算，尺寸一致
a > b
```

```python
c = pd.Series(np.arange(4))

# 不同维度，广播运算，默认在轴axis=1（行）
a > c
```
