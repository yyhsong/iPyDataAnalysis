```python
import numpy as np
import matplotlib.pyplot as plt

# 原始数据
X = [1, 2, 3, 4, 5, 6]
Y = [2.6, 3.4, 4.7, 5.5, 6.47, 7.8]

# 用一次多项式拟合，相当于线性拟合
z = np.polyfit(X, Y, 1)
# p = np.poly1d(z)

# 绘图显示
x = np.arange(1, 8)
y = z[0] * x + z[1]

plt.scatter(X, Y)
plt.plot(x, y, marker='o', mfc='r')

plt.show()
```
