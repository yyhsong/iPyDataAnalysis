```python
import matplotlib
import matplotlib.pyplot as plt
```

# pyplot默认不支持中文显示，需要rcParams修改字体

rcParams的属性
- font.family 字体
- font.style 字体风格，正常normal或斜体italic
- font.size 字体大小，整数或者large, x-small

常用中文字体
- SimSun 宋体
- SimHei 黑体
- Kaiti 楷体
- STSong 华文宋体
- Microsoft YaHei 微软雅黑 

```python
matplotlib.rcParams['font.family'] = 'Microsoft YaHei'
matplotlib.rcParams['font.size'] = 14

plt.plot([3, 1, 4, 5, 2])
plt.ylabel('纵轴（值）')
plt.show()
```

```python
# 显示中文的第二种方法，在有中文输出的地方增加fontproperties属性

plt.plot(['苹果', '香蕉', '蜜桃', '桔子', '梨'], [3, 1, 4, 5, 2])
plt.xlabel('横轴（分类）', fontproperties='SimHei', fontsize=14)
plt.ylabel('纵轴（值）', fontproperties='SimHei', fontsize=14)
plt.show()
```
