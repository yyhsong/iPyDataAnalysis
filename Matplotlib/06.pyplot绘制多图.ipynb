# 绘制多图

使用 pyplot 中的 subplot() 和 subplots() 方法来绘制多个子图，subplot() 方法在绘图时需要指定位置，subplots() 方法可以一次生成多个，在调用时只需要调用生成对象的 ax 即可

plt.subplot(nrows, ncols, plot_number)

在全局绘图区域中创建一个分区体系，并定位到一个子绘图区域

```python
import numpy as np

def f(n):
    return np.exp(-n) * np.cos(2*np.pi*n)

a = np.arange(0.0, 5.0, 0.02)

plt.subplot(2, 1, 1)  # 构建2行1列的分区体系，并使用第1个绘图区域
plt.plot(a, f(a))

plt.subplot(212)  # 等价于 subplot(2, 1, 2)
plt.plot(a, np.cos(2*np.pi*a), 'r--')

plt.show()
```

# 复杂绘图区域

# plt.subplot2grid(GridSpec, CurSpec, colspan=1, rowspan=1)

设定网格，选择绘图区域，确定选中行列数量，编号从0开始

- plt.subplot2grid((3,3), (0,0))  3行3列的Grid，选择第1行第1列plot
- plt.subplot2grid((3,3), (1,0), colspan=2)  3行3列的Grid，选择第2行第1列的plot，该plot占两列
- plt.subplot2grid((3,3), (1,2), rowspan=2)  3行3列的Grid，选择第2行第3列的plot，该plot占两行

```python
# GridSpec类

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

gs = gridspec.GridSpec(3, 3)  # 3行3列的Grid

plot1 = plt.subplot(gs[0, :])  # 第1行第1列，该区域占3列
plot2 = plt.subplot(gs[1, :-1]) # 第2行第1列，该区域占2列
plot3 = plt.subplot(gs[1:, -1])  # 第2行第3列，该区域占两行
plot4 = plt.subplot(gs[2, 0])  # 第3行第1列
plot5 = plt.subplot(gs[2, 1])  # 第3行第2列
```
