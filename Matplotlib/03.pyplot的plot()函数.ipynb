```python
import numpy as np
import matplotlib.pyplot as plt
```

# plt.plot(x, y, format_string, **kwargs)

- x x轴数据，列表或数组，可选
- y y轴数据，列表或数组
- format_string 控制曲线的格式字符串，可选
- **kwargs 第二组或更多(x, y, format_string)

当绘制多条曲线时，各条曲线的x不能省略

```python
a = np.arange(10)
plt.plot(a, a*1, a, a*2, a, a*3, a, a*4)  # 每两个值为一组
plt.show()
```

# format_string 控制曲线的格式字符串，由颜色字符、风格字符和标记字符组成

颜色字符
- r 红色
- g 绿色
- b 蓝色
- y 黄色
- k 黑色
- w 白色
- m 洋红色 magenta
- c 青绿色 cyan
- #00ff00 rgb某颜色
- 0.8 灰度值字符串

风格字符
- \- 实线
- \-- 破折线
- \-. 点划线
- : 虚线
- '' 无线条

标记字符
- . 点标记
- \* 星形标记
- , 像素标记（极小点）
- o 实心圆圈标记
- s 实心方形标记
- p 实心五角标记
- h 竖六边形标记
- H 横六边形标记
- \+ 十字标记
- x x标记
- D 菱形标记
- d 瘦菱形标记
- \| 垂直线标记
- v 倒三角标记
- ^ 上三角标记
- < 左三角标记
- \> 右三角标记
- 1 下花三角标记
- 2 上花三角标记
- 3 左花三角标记
- 4 右花三角标记

颜色字符、风格字符和标记字符可以组合使用

```python
a = np.arange(10)
plt.plot(a, a*1, 'r-o', a, a*2, 'g--*', a, a*3, 'b-.s', a, a*4, 'y:p')
plt.show()
```

# **kwargs 第二组或更多(x, y, format_string)

- color 颜色 color='red'
- lineStyle 线条风格 lineStyle='dashed'
- marker 标记风格 marker='o'
- markerfacecolor 标记颜色 markerfacecolor='blue'
- markersize 标记尺寸 markersize=10

```python
# 设置图例
a = np.arange(10)
for i in range(1, 4):
    plt.plot(a, i*a**2, label='Line %d' % i)
    
plt.legend(loc='best')
plt.show()
```
