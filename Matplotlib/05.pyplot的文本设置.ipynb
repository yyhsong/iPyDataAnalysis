```python
import numpy as np
import matplotlib.pyplot as plt
```

# pyplot的文本设置函数

- plt.xlabel() x轴名称
- plt.ylabel() y轴名称
- plt.title() 图表名称
- plt.text() 在任意位置增加文本
- plt.annotate() 在图形中增加带箭头的注解

```python
a = np.arange(0.0, 5.0, 0.02)

plt.plot(a, np.cos(2*np.pi*a), 'r--')

plt.title(r'正弦波 $y=cos(2\pi x)$', fontproperties='SimHei', fontsize=18)
plt.xlabel('横轴（时间）', fontproperties='SimHei', fontsize=14, color='blue')
plt.ylabel('纵轴（振幅）', fontproperties='SimHei', fontsize=14, color='green')

#plt.text(2, 1, r'$\mu=100$', fontsize=12)
plt.annotate(r'$\mu=100$', xy=(2, 1), xytext=(3, 1.5), arrowprops=dict(facecolor='black', shrink=0.1, width=1))

plt.axis([-1, 6, -2, 2])
plt.grid(True)
plt.show()
```
