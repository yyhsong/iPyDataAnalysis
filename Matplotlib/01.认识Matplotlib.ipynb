# Matplotlib: Visualization with Python

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

- Matplotlib库由各种可视化类构成
- pyplot是绘制各类可视化图形的命令子库

```python
import matplotlib.pyplot as plt
```

```python
# 当plot()方法只有一个列表参数时，被做为y轴的值
# x轴以索引自动生成
plt.plot([2, 4, 3, 1, 5])
plt.show()
```

```python
# savefig()将输出图形存储为文件，默认PNG格式
# 可以通过dpi修改输出质量
plt.plot([2, 4, 3, 1, 5])
plt.ylabel('Grade')
plt.savefig('test', dpi=600)  # PNG图片
plt.show()
```

```python
# 当plot()方法有两个参数时，第一个为x轴数据，第二个为y轴数据
plt.plot([1, 3, 5, 7, 9], [2, 4, 3, 1, 5])
plt.ylabel('Grade')
plt.axis([0, 10, 0, 6])  # 前两个值为x轴的起始值，后两个值为y轴的起始值
plt.show()
```
