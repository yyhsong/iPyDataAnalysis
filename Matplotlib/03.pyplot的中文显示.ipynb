```python
import matplotlib
import matplotlib.pyplot as plt
```

# pyplot默认不支持中文显示，需要rcParams修改字体

rcParams的属性
- font.family 字体
- font.style 字体风格，正常normal或斜体italic
- font.size 字体大小，整数或者large, x-small

常用中文字体
- SimSun 宋体
- SimHei 黑体
- Kaiti 楷体
- STSong 华文宋体
- Microsoft YaHei 微软雅黑 

```python
matplotlib.rcParams['font.family'] = 'Microsoft YaHei'
matplotlib.rcParams['font.size'] = 12
```

# 标题和标签的定位

- title() 方法提供了 loc 参数来设置标题显示的位置，可以设置为: 'left', 'right', 和 'center'， 默认值为 'center'
- xlabel() 方法提供了 loc 参数来设置 x 轴显示的位置，可以设置为: 'left', 'right', 和 'center'， 默认值为 'center'
- ylabel() 方法提供了 loc 参数来设置 y 轴显示的位置，可以设置为: 'bottom', 'top', 和 'center'， 默认值为 'center

```python
# 显示中文的第二种方法，在有中文输出的地方增加fontproperties属性

plt.title('水果派', fontproperties='SimHei', fontsize=14)

plt.plot(['苹果', '香蕉', '蜜桃', '桔子', '梨'], [3, 1, 4, 5, 2])
plt.xlabel('水果', fontproperties='SimHei', fontsize=12, loc='right')
plt.ylabel('战斗力', fontproperties='SimHei', fontsize=12, loc='top')
plt.show()
```
